Plotting the categorical features in bar charts.

```python
import plotly.express as px

#A seguinte função vai plotar a quantidade de Ids transportadas em função do dataframe feature de interesse
def plot_qtd_transported(df, feature):
    fig = px.bar(df.groupby([feature, 'Transported']).count()['PassengerId'].reset_index(),
                 x='Transported', y='PassengerId', facet_col=feature, color='Transported',
                 color_continuous_scale=['red', 'green'], template='plotly_dark')
    fig.update_coloraxes(showscale=False)
    #fig.layout.update(showlegend=False)
    fig.show(theme='dark')

plot_qtd_transported(df, 'HomePlanet')
plot_qtd_transported(df, 'CryoSleep')
plot_qtd_transported(df, 'Destination')
plot_qtd_transported(df, 'VIP')
plot_qtd_transported(df, 'Cabin_deck')
plot_qtd_transported(df, 'Cabin_side')
```

## Normalization and exclusion of features.

* Features related to strings and Ids will be removed.
* Categorical features will be expanded by the get_dummies() method.
* The only continuous feature, Age, will undergo a standardization and scaling process.

```python
df_cryo = df_cryo.drop(['CryoSleep', 'PassengerId', 'Cabin', 'Name'], axis=1)
df_cryo
```

```python
df_cryo = pd.get_dummies(df_cryo)
df_cryo
```

```python
def normalize(s):
    #StandardScaler
    mean = s.mean()
    std = s.std()
    s = (s-mean)/std

    #MinMaxScaler
    max = s.max()
    min = s.min()
    s = (s-min)/(max-min)
    return s

df_cryo['Age'] = normalize(df_cryo['Age'])
df_cryo
```

```python
df_not_cryo = df_not_cryo.drop(['CryoSleep', 'PassengerId', 'Cabin', 'Name'], axis=1)
df_not_cryo = pd.get_dummies(df_not_cryo)

df_not_cryo['Age'] = normalize(df_not_cryo['Age'])
df_not_cryo['RoomService'] = normalize(df_not_cryo['RoomService'])
df_not_cryo['FoodCourt'] = normalize(df_not_cryo['FoodCourt'])
df_not_cryo['ShoppingMall'] = normalize(df_not_cryo['ShoppingMall'])
df_not_cryo['Spa'] = normalize(df_not_cryo['Spa'])
df_not_cryo['VRDeck'] = normalize(df_not_cryo['VRDeck'])

df_not_cryo
```

# Application of Machine Learning.

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
random_state = 0

def apply_cv(df, model, random_state=0):
    X = df.drop('Transported', axis=1)
    y = df['Transported']

    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=random_state)
    scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
    return scores

from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier()

apply_cv(df_cryo, neigh)
```

```python
def iterating_k(df, step=1, max=len(df)):
    k_list = list(range(1, max, step))
    mean_list = []
    std_list = []

    for k in k_list:
        scores = apply_cv(df, KNeighborsClassifier(algorithm='brute', n_neighbors=k))
        mean_list.append(scores.mean())
        std_list.append(scores.std())

    return pd.DataFrame({'k': k_list, 'mean': mean_list})

df_cv = iterating_k(df_cryo, step=1, max=500)
px.scatter(df_cv, x='k', y='mean', template='plotly_dark').show()
```

```python
df_cv[df_cv['mean'] == df_cv['mean'].max()]
```

```python
df_cv = iterating_k(df_not_cryo, step=1, max=500)
px.scatter(df_cv, x='k', y='mean', template='plotly_dark').show()
```

```python
df_cv[df_cv['mean'] == df_cv['mean'].max()]
```
